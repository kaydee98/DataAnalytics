### Research Questions ###
1. What demographic factors are most associated with trip duration?

### The Dataset ###
This dataset includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. 
The dataset contains 183,412 entries with 16 columns. Each ride is represented by a single row in the dataset. The dataset contains information on bike-share trips and details of each trip's duration, start and end times, and station locations (start and end stations). It also includes user demographics, such as birth year, gender, and user type.

### Structure of the dataset?

Below is an overview of the structure of the dataset:

- **bike_id**: Unique identifier for the bike used (integer)

- **start_time**: Timestamps for when the trip started (string)
- **end_time**: Timestamps for when the trip ended (string)
- **start_station_id**: ID of the start station (float)
- **start_station_name**: Names of the start station (string)
- **start_station_latitude**: Geographic coordinates for the start station - Lat (float)
- **start_station_longitude**: Geographic coordinates for the start station - Long (float)
- **end_station_id**: IDs of the end stations (float)
- **end_station_name**: Names of the end station (string)
- **end_station_latitude**: Geographic coordinates for the end station - Lat(float)
- **end_station_longitude**: Geographic coordinates for the end station - Long (float)
- **user_type**: Type of user (string)
- **member_birth_year**: Birth year of the user (float).
- **member_gender**: Gender of the user (string).
- **bike_share_for_all_trip**: Indicates if the bike is part of an all-trip share program (string).

## Univariate Exploration
In this section, I investigate the distributions of individual variables, taking a deeper look at unusual points or outliers, to clean things up and prepare to look at the relationships between variables. 

```python
# Create and show histogram
def show_histogram(df, title, xlabel, ylabel):
    plt.hist(df, color='skyblue')  
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title); 
    plt.show();
```

```python
# Histogram of trip duration
ford_gobike_data['duration_minute'] = ford_gobike_data['duration_sec']/60  # Converting seconds to minutes
show_histogram(ford_gobike_data['duration_minute'], 'Distribution of Trip Duration', 'Trip Duration (minutes)', 'Frequency')
```

```python
## Bar Chart for User Types
user_type_counts = ford_gobike_data['user_type'].value_counts()
plt.bar(user_type_counts.index, user_type_counts.values, color='orange')
plt.xlabel('User Type')
plt.ylabel('Count')
plt.title('Count of User Types');
plt.show();
```

```python
sns.countplot(data=ford_gobike_data, x='member_gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution Of Bike-Sharing Users');
plt.show();
```

```python
# Compute member age
current_year = 2019 #Year the dataset was gathered
ford_gobike_data['member_age'] = (current_year - ford_gobike_data['member_birth_year'])

# Histogram of Age distribution
show_histogram(ford_gobike_data['member_age'], 'Distribution of Member Age', 'Age in years', 'Frequency')
```

### Discussion of the distribution(s) of variables of interest 

1. **Trip Duration (Histogram)**: The distribution of trip duration, visualized through a histogram, is right-skewed, with a high frequency of short trips with a tapering number of longer trips. This pattern suggests that most bike-share users take shorter trips. Converting the trip duration to minutes and using a log of the data my revealed better information about the distribution pattern. A transformation was done to convert trip duration from seconds to minutes to improve interpretability

2. **User Type (Bar Chart)**: The distribution of user_type shows a majority of the entries falling under the subscriber category. This suggests that the bike-share system has a strong core user base who use the service regularly. Casual users make up a smaller portion of the dataset. 

3. **Gender Distribution (Count Plot)**: The gender distribution is also uneven, with a majority of users identifying as Male, followed by Female users. The Other category, which includes unknown or non-disclosed gender entries, is the smallest group. This pattern suggests that the service may be more popular among male users. The small number of users in the "Other" category may indicate incomplete data entry or a lack of options for non-binary or undisclosed users
  
5. **Age Distribution (Histogram)**: The histogram of member age distribution is highly right-skewed, with the majority of users concentrated between 20 and 40 years old, and a sharp decline in frequency as age increases. Few users are above 70 suggesting limited adoption among older demographics. Additionally, improbable values above 100 years indicate potential data entry errors, highlighting a need for data cleaning.

## Bivariate Exploration

In this section I will investigate the relationships between pairs of variables in the dataset.

```python
## Scatter plot of Trip Duration and Birth Year
sns.scatterplot(data=ford_gobike_data, x=ford_gobike_data['member_age'], y=ford_gobike_data['duration_minute'], alpha=0.5)
plt.title('Trip Duration vs. Age')
plt.xlabel('Age')
plt.ylabel('Trip Duration (mins)')
plt.show()
```

```python
#Relationship between Trip Duration and Gender - Box Plot
sample_size = 6000
ford_gobike_sample = ford_gobike_data.sample(n=sample_size, random_state=8) 

sns.boxplot(data=ford_gobike_sample[ford_gobike_sample['member_gender'].notna()], 
            x='member_gender', y=ford_gobike_sample['duration_minute'])
plt.title('Trip Duration by Gender')
plt.xlabel('Gender')
plt.ylabel('Trip Duration (mins)')
plt.show()
```

```python
# Creating a clustered bar chart of gender distribution by user type
sns.countplot(data=ford_gobike_data, x='user_type', hue='member_gender')
plt.title("Gender Distribution by User Type")
plt.xlabel("User Type")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

### Discussion of the relationships observed in the Bivariant Exploration
Based on the bivariate exploration of *trip duration* in relation to *birth year (age)* and *gender*, the following relationships and variations were observed:

1. **Trip Duration and Age (Scatter Plot)**_
Trip duration tends to be more variable and potentially longer as age increases, particularly for users aged 20-60. Most trips across all ages cluster within 0-200 minutes. Younger users (20-30 years) generally take shorter, consistent trips, while older users (60+ years) exhibit a gradual decline in both trip duration and variability. There appears to be a potential quality issues with member age over 80 years which seems improbable. In addition, large number of outliers in the trip duration, trips 100x over the mean of the dataset, which could be the result of data collection issues or bikes been abandoned.

2. **Trip Duration and Gender (Box Plot)**
The box plot of trip duration by gender reveals that all three categories have similarly skewed distributions, with the majority of trips concentrated in shorter durations and numerous outliers in higher ranges. The median trip duration is similar across genders, clustered around 10 minutes. Male users show the most extreme outliers, with some trips reaching around 1200 minutes, while Female users have outliers up to about 800 minutes. Overall, despite differences in outliers and sample size, the main pattern of shorter trip durations is consistent across all gender categories.

3. **Gender Distribution by User Type (Clustered Bar Plot)**
The scatter plot of the gender distribution by user type indicates notable differences in the distribution across user types. Females and users labeled as "Other" make up a smaller portion of both groups but are relatively more evenly distributed within the "Customer" category compared to "Subscriber."

## Multivariate Exploration
Finally we take a look at the relationships between three categories of features 

```python
# Creating a FacetGrid to analyze trip duration by user type, gender and age
g = sns.FacetGrid(ford_gobike_data[ford_gobike_data['member_gender'].notna()], col='user_type', row='member_gender', 
                  margin_titles=True, height=4, aspect=1.2)
g.map(sns.scatterplot, 'member_age', 'duration_minute')
g.set_axis_labels('Age', 'Trip Duration (min)')
g.set_titles(col_template='{col_name}', row_template='{row_name}')
g.add_legend()
plt.show()
```

```python
# Creating a scatter plot of trip duration by age with customer type as color encoding
sns.scatterplot(data=ford_gobike_data, x='member_age', y='duration_minute', hue='user_type', alpha=0.6)
plt.title("Trip Duration by Age and User Type")
plt.xlabel("Age")
plt.ylabel("Trip Duration (minutes)")
plt.legend(title="User Type")
plt.show()
```

### Discussion of the relationships observed in the Multivariate Exploration
Based on the Multivariate exploration of *trip duration*, *user type*, *birth year (age)* and *gender*, the following relationships and variations were observed

1. **FacetGrid Plot**The plot shows the relationship between trip duration and age, segmented by user type (Customer vs. Subscriber) and gender (Male, Female, Other). Subscribers tend to have more consistent trip durations across age groups, while Customers show slightly more variability. Most users are in the 20-60 age range across all categories. Trip durations generally cluster below 200 minutes, though longer trips (over 800 minutes) are more common among users aged 30-60. Overall, trip duration are relatively consistent across genders, but user type seems to influence trip duration more than gender does.
2. **Trip Duration by Age & User Type (Clustered Bar Plot)**
This visualization reveals how trip duration varies across different age groups and user types. The trip patterns is largely similar between the user types across ages with lower trip duration with age
